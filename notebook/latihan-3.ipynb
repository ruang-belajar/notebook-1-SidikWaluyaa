```python
# PERSIAPAN DATA

# download file contoh csv
!wget https://raw.githubusercontent.com/ruang-belajar/python/master/notebook/data/fashion.csv -O fashion.csv


import pandas as pd

# load data ke dataframe (df)
df = pd.read_csv("fashion.csv", sep=';')
df.head()
```

Tulis perintah untuk menampilkan data dengan Shipping VIP dan Priority High.


```python
df1 = df[(df['Shipping'] == 'VIP') & (df['Priority'] == 'High')]
df1
```

Profit perbulan

```python
df2 = df.groupby('Month')['Profit'].sum()
df2
```

Tulis perintah untuk menampilkan data dengan Shipping Economy.


```python
df3 = df[df['Shipping'] == 'Economy']
df3
```

Tulis perintah untuk menampilkan rata-rata Profit per Category (menggunakan groupby)

```python
df4=df.groupby('Category')['Profit'].mean()
df4
```
